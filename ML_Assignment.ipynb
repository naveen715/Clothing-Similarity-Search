```python
#Step-01: Prepare Dataset 
import pandas as pd
import requests
from bs4 import BeautifulSoup

data = []
URL = 'https://www.boohoo.com/mens/shirts?menshome_mens-shirts'
page_number = 1
counter = 0

while counter < 1000:
    page_url = f'{URL}&page={page_number}'
    page = requests.get(page_url)
    soup = BeautifulSoup(page.content, 'html.parser')
    product_tiles = soup.find_all(attrs={'class': 'b-product_tile'})

    if not product_tiles:
        break

    if page.status_code == 200:
        print(f"Fetching data from page {page_number}")
        for tile in product_tiles:
            if counter >= 1000:
                break
            title = tile.find(attrs={'class': 'b-product_tile-title'})
            price = tile.find(attrs={'class': 'b-price-item'})
            url = tile.find('a')['href']  # Extract the URL from the 'href' attribute of the <a> tag
            if title and price:
                data.append([url, title.text.strip(), price.text.strip()])
                counter += 1

        page_number += 1
    else:
        print("URL Not Found")
        break

df = pd.DataFrame(data, columns=['URL', 'Title', 'Price'])
df.to_csv('clothes_materials.csv')
```

```python
#Step-02: Preprocess and cleansing of data 

import pandas as pd
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

# Load the data from the CSV file
df = pd.read_csv('/content/combined.csv')  

# Preprocess the 'Title' column
stopwords = set(stopwords.words('english'))
stemmer = PorterStemmer()

def preprocess_text(text):
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Convert text to lowercase
    text = text.lower()
    
    # Tokenize the text
    tokens = word_tokenize(text)
    
    # Remove stop words
    tokens = [token for token in tokens if token not in stopwords]
    
    # Perform stemming
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    
    # Join the tokens back into a single string
    preprocessed_text = ' '.join(stemmed_tokens)
    
    return preprocessed_text

df['Title'] = df['Title'].apply(preprocess_text)

# Save the preprocessed data to a new CSV file
df.to_csv('preprocessed_combined_items.csv',index=False) 
```

```python
#Step-3: Extract the features and compute the similarity score
#Function to extract features and compute the similarity index
import pandas as pd                                           
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load the data from the preprocessed CSV file
df = pd.read_csv('/content/preprocessed_combined_items.csv')  

# Extract features using TF-IDF
vectorizer = TfidfVectorizer()
features = vectorizer.fit_transform(df['Title'])

# Compute similarity between input text and database text
input_text = "short trench coat"
input_features = vectorizer.transform([input_text])

similarity_scores = cosine_similarity(input_features, features).flatten()

# Sort the similarity scores in descending order and retrieve the indices
similarity_indices = similarity_scores.argsort()[::-1]

# Print the most similar texts
print("Most similar texts:")
for idx in similarity_indices:
    similarity_score = similarity_scores[idx]
    if similarity_score > 0:
        print(df['Title'][idx], "- Similarity Score:", similarity_score)
    else:
        break
```

```python
#Final function which takes a text of required item and number of results required as input and returns the product link 
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import json

# Step-4: Return the ranked results URLS of top similar items
def get_top_similar_urls(input_text, df):
    # Extract features using TF-IDF
    vectorizer = TfidfVectorizer()
    features = vectorizer.fit_transform(df['Title'])

    # Compute similarity between input text and database text
    input_features = vectorizer.transform([input_text])
    similarity_scores = cosine_similarity(input_features, features).flatten()

    # Sort the similarity scores in descending order and retrieve the indices
    similarity_indices = similarity_scores.argsort()[::-1]

    # Get the URLs of the top n most similar items
    print("Enter a value")
    n = int(input())
    top_similar_urls = []
    count = 0
    for idx in similarity_indices:
        similarity_score = similarity_scores[idx]
        if similarity_score > 0 and df['URL'][idx] not in top_similar_urls:
            top_similar_urls.append(df['URL'][idx])
            count += 1
            if count >= n:
                break

    return top_similar_urls

# Load the preprocessed data into a DataFrame
df = pd.read_csv("/content/preprocessed_combined_items.csv") 
#Tell user to enter the text name 
print("Enter input text item name") 
input_text=input() 
# Call the function to get the top similar URLs
top_url_json = []
top_urls = get_top_similar_urls(input_text, df)
for url in top_urls:
  if not url.startswith('https://www.boohoo.com'):url = 'https://www.boohoo.com' + url
  top_url_json.append(url)
# Print the top similar URLs
print("Top similar URLs:")
if(len(top_urls)==0):
  print("Sorry no entry in databse")
else:
  for url in top_url_json:
    print(url)
print("JSON Response:")
response = {'input_text': input_text, 'top_urls': top_url_json}
print(json.dumps(response))
```
